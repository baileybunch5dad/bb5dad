```python
# https://www.youtube.com/watch?v=UpS9VdY7Nxg
import numpy as np
# 14 not blue
# 16 not yellow
# 24 not red
# 12 not pink
# 0b + 1y + 1r + 1p = 14
# 1b + 0y + 1r + 1p = 16
# 1b + 1y + 0r + 1p = 24
# 1b + 1y + 1r + 0p = 12
a = np.array([[0,1,1,1],[1,0,1,1],[1,1,0,1],[1,1,1,0]])
b = np.array([[14],[16],[24],[12]])
x = np.dot(np.linalg.inv(a),b)
print(a)
print(b)
print(x)
```

```python
from mpl_toolkits import mplot3d

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


# Plot 3 planess in 3d
# Equation
#
#    x + y + 2*z = 9 ==> z=(9-x-y)/2
#   2x - y - z = 4   ==> z=2x-y-4
#   -x + 5y +2z = 7  ==> z = (7+x-5*y)/2

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x, y = np.meshgrid(range(10), range(10))
z = (9 - x - y) / 2
ax.plot_surface(x, y, z, alpha=0.5)
z = 2 * x - y - 4
ax.plot_surface(x, y, z, alpha=0.5)
z = (7 + 4*x - 6 * y) / 2
ax.plot_surface(x, y, z, alpha=0.5)
plt.show()



```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

n = 100
for h in np.linspace(.01,3,5):
  theta = np.linspace(0, 2*np.pi, n)
  r = np.linspace(0,np.sqrt(h),100)
  x = np.outer(r, np.cos(theta))
  y = np.outer(r, np.sin(theta))
  z = np.repeat(h, n*n).reshape((n,n))
  ax.plot_surface(x,y,z)

plt.show()
```

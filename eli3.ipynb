```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

fig, ax = plt.subplots()

x = np.linspace(0, 2*np.pi, 500)
line1, = ax.plot(x, np.sin(x))
print('line1',line1)
l2x = np.array([0])
l2y = np.sin(l2x)
line2, = ax.plot(l2x, l2y, 'ro')
lines = list([line1,line2]) 


def animate(i):
    tl2x = np.array([x[i % 500]])
    tl2y = np.sin(tl2x)
    line2.set_xdata(tl2x)
    line2.set_ydata(tl2y)
    return lines


ani = animation.FuncAnimation(
    fig, animate, interval=20, blit=True, save_count=500)

# To save the animation, use e.g.
#
# ani.save("movie.mp4")
#
# or
#
# writer = animation.FFMpegWriter(
#     fps=15, metadata=dict(artist='Me'), bitrate=1800)
# ani.save("movie.mp4", writer=writer)

# f = r"c://Users/chris/animation.mp4"
# writervideo = animation.FFMpegWriter(fps=60)
# ani.save(f, writer=writervideo)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
# graph 1/x from near 0 
x = np.linspace(.2, 2, 100)
y = 1/x
plt.plot(x,y)
plt.grid()
# show the slope at .4
x1 = .4
y1 = 1/x1
m = -1/(x1*x1)
# (y2-y1)/(x2-x1)=m
# m(x2-x1)=y2-y1
# since y2=0 m(x2-x1)=-y1
# x2-x1=-y1/m
# x2=x1-y1/m
y2=0
x2=x1-y1/m
plt.plot([x1,x2],[y1,y2],'-')
plt.title('Slope at '+str(x1)+'='+str(m))
plt.show()
```

## Prove $\frac{d}{dx} \frac{1}{x} = -\frac{1}{x^2}$

Proof

$$
\begin{aligned}
\frac{d}{dx} \frac{1}{x} &= \lim_{\Delta x \to 0} \frac{\frac{1}{x + \Delta x}-\frac{1}{x}}{\Delta x} & \text{to prove} \\
&= \lim_{\Delta x \to 0} \frac{1}{\Delta x} \left( \frac{x - (x+\Delta x)}{x(x+\Delta x)} \right) & \text{pull out }\frac{1}{\Delta x}\text{mult by }x(x+\Delta x)\\
&= \lim_{\Delta x \to 0}\frac{1}{\Delta x} \left( \frac{-\Delta x}{x(x+\Delta x)} \right) & x-x=0\\
&=  \lim_{\Delta x \to 0} \frac{-1}{x(x+\Delta x)} & \frac{\Delta x}{\Delta x} \text{cancels}\\
&=   -\frac{1}{x^2} & \Delta x \to 0 \\
\end{aligned}
$$
